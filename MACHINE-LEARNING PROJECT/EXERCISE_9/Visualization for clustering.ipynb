```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=1.5)

plt.figure(figsize=(6,4))
plt.scatter(X[:,0], X[:,1], c='blue', alpha=0.6, edgecolors='w')
plt.title("Basic Scatterplot (Matplotlib)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df = pd.DataFrame(X, columns=['Feature1','Feature2'])
df['Cluster'] = y

sns.scatterplot(data=df, x="Feature1", y="Feature2", hue="Cluster", palette="deep")
plt.title("Seaborn Scatterplot with Cluster Coloring")
plt.show()
```

```python
df.plot.scatter(x="Feature1", y="Feature2", c="Cluster", colormap="viridis", alpha=0.7)
plt.title("Scatterplot via Pandas DataFrame")
plt.show()
```

```python

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 170, 160, 180, 200]      
expenses = [100, 120, 130, 140, 150, 160]  

plt.figure(figsize=(8,5))
plt.bar(months, sales, color='skyblue', label="Sales")
plt.bar(months, expenses, color='salmon', alpha=0.7, label="Expenses")
plt.title("Monthly Sales vs Expenses - Bar Chart")
plt.xlabel("Months")
plt.ylabel("Amount (in $1000s)")
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o', label="Sales", linewidth=2)
plt.plot(months, expenses, marker='s', label="Expenses", linewidth=2)
plt.title("Monthly Sales vs Expenses - Line Plot")
plt.xlabel("Months")
plt.ylabel("Amount (in $1000s)")
plt.legend()
plt.grid(True)
plt.show()
```
