```python
#load the bird migration dataset
bird_migration_data = pd.read_csv('bird_migration_data.csv')
# Display the first few rows of the dataset
print(bird_migration_data.head())
# Display the data types of each column
print(bird_migration_data.dtypes)
```

```python
#filter the dataset to include only birds that are goose species 
goose_species = bird_migration_data[bird_migration_data['Species'].str.contains('Goose')]
```

```python
# count of goose species by region
goose_counts_by_region = goose_species['Region'].value_counts()
# Display the count of goose species by region
print(goose_counts_by_region)
```

```python
# line plot showing average resting blood pressure across age groups

# group data by age
Duration = bird_migration_data.groupby('Region')['Flight_Distance_km'].mean().reset_index()

# Plot
sns.lineplot(data=Duration, x='Region', y='Flight_Distance_km', color='teal')
plt.title('Average Flight Distance Across Region')
plt.xlabel('Region')
plt.ylabel('Avg Flight Distance (km)')

# Display the plot
plt.show()


```

```python
# line plot of migration reason by region
sns.scatterplot(data=bird_migration_data, x='Predator_Sightings', y='Temperature_C')
plt.title('Predator_Sightings by Temperature_C')
plt.xlabel('Predator Sightings')
plt.ylabel('Temperature (Celsius)')
plt.figure(figsize=(15, 9))
plt.show()
```

```python
#bar chart of goose species by region
plt.figure(figsize=(10, 6))
sns.countplot(data=goose_species, x='Region', palette='viridis')
plt.title('Goose Species Count by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
category_counts = bird_migration_data['Food_Supply_Level'].value_counts()

# Pie Chart of location
category_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8, 8),
    startangle=90,
    colors=plt.cm.Paired.colors
)
plt.title("Pie Chart Of Food Supply Level")
plt.ylabel("")
plt.show()
```
